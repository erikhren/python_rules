# Lists

**List is a sequence of values (like a string)** <br>
- characters can be any type
- Create a list --> enclose in [el1,el2...]
- ['spam', 2.0, 5, [10, 20]] --> string, float, int, another list (NESTED LIST)
- List can be assigned to variables

**Lists are MUTABLE**
- same as accessing a string char --> list[0]
- can change and order OR reassign an item in a list
- When the bracket operator appears on the left side of an assignment, it identifies the element of the list that will be assigned.<br>
numbers = [17, 123]<br>
numbers[1] = 5<br>
print(numbers)<br>
[17, 5]<br>
<br>
-  list is like a relationship between indices and elements.
- This relationship is called a mapping; each index “maps to” one of the elements
- indexes work the same as string indexes (any integer, if not there--> error, -1 --> backwards)
- IN (in) operator works the same --> True/False (if in the list)
<br>
**Traversing a list**
- for loop (same syntax)
- To read:<br>
for cheese in cheeses:<br>
print(cheese)<br>
<br>
- To write/update elements --> need indexes (i.e. use range(), len()):<br>
for i in range(len(numbers)):<br>
numbers[i] = numbers[i] * 2<br>
<br>
- This loop traverses the list and updates each element. 
- len returns the number of elements in the list. 
- range returns a list of indices from 0 to n − 1, where n is the length of the list. 
- Each time through the loop, i gets the index of the next element. 
- The assignment statement in the body uses i to read the old value of the element and to assign the new value. A for loop over an empty list never executes the body:<br>
for x in empty:<br>
print('This never happens.')<br>
<br>

**Aliasing**
- If a refers to an object and you assign b = a, then both variables refer to the same object:<br>
a = [1, 2, 3]<br>
b = a<br>
b is a<br>
True<br>
<br>
- The association of a variable with an object is called a reference (here, two references to the same object).
- An object with more than one reference has more than one name, so we say that the object is aliased.
- If the aliased object is mutable, changes made with one alias affect the other:<br>
b[0] = 17<br>
print(a)<br>
[17, 2, 3]<br>
- Can be useful BUT errors can occur easier --> AVOID ALIASING with mutable objects (i.e. lists, dicts...)
- For immutable objects like strings, aliasing is not as much of a problem. In this example:<br>
a = 'banana'<br>
b = 'banana'<br>
<br>
**List Arguments**
- When you pass a list to a function, the function gets a reference to the list. 
- If the function modifies a list parameter, the caller sees the change. 
- For example, delete_head removes the first element from a list:<br>
def delete_head(t):<br>
del t[0]<br>
<br>
- Here’s how it is used:<br>
letters = ['a', 'b', 'c']<br>
delete_head(letters)<br>
print(letters)<br>
['b', 'c']<br>
- The parameter t and the variable letters are ALIASES for the same object.
<br>
- It is important to distinguish between operations that modify lists and operations that create new lists.<br>
- For example, the append method modifies a list, but the + operator creates a new list:<br>
t1 = [1, 2]<br>
t2 = t1.append(3)<br>
print(t1)<br>
[1, 2, 3]<br>
print(t2)<br>
None<br>
<br>
t3 = t1 + [3]<br>
print(t3)<br>
[1, 2, 3]<br>
t2 is t3<br>
False<br>
<br>
- Important when you write functions that are supposed to modify lists. <br>
- For example, this function does not delete the head of a list:<br>
def bad_delete_head(t):<br>
t = t[1:] # WRONG!<br>
<br>
- The slice operator creates a new list and the assignment makes t refer to it, but none of that has any effect on the list that was passed as an argument.<br>
<br>
- An alternative is to write a function that creates and returns a new list. <br>
- For example, tail returns all but the first element of a list:<br>
def tail(t):<br>
return t[1:]<br>
This function leaves the original list unmodified. Here’s how it is used:<br>
letters = ['a', 'b', 'c']<br>
rest = tail(letters)<br>
print(rest)<br>
['b', 'c']<br>
