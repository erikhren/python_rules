#### Random Numbers (Many random modules/fun. available)

- Given the same inputs, most computer programs generate the same outputs every time, so they are said to be DETERMINISTIC (good for calculations).
- For some applications, though, we want the computer to be unpredictable --> Games
- there are ways to make it at least seem nondeterministic: 
- use algorithms that generate pseudorandom numbers (not truly random because they are generated by a deterministic computation, just looking is impossible to distinguish if they are not).
- The RANDOM module provides functions that generate pseudorandom numbers (which I will simply call “random” from here on).
- The function random returns a random float between 0.0 and 1.0 (including 0.0 but not 1.0). 
- Each time you call random, you get the next number in a long series.
<br>

##### To see a sample, run this loop:

import random<br>
for i in range(10):<br>
x = random.random()<br>
print(x)<br>
This program produces the following list of 10 random numbers between 0.0 and<br>
up to but not including 1.0.<br>
<br>
- The function RANDOM.RANDINT(low,high) takes the parameters low and high, and returns an integer between low and high (including both).
<br>
- To choose an element from a sequence at random, you can use choice: random.choice(t)
<br>
- The random module also provides functions to generate random values from 
continuous distributions including Gaussian, exponential, gamma, and a few more.
<br>

#### ADDING NEW FUNCTIONS

- A function definition specifies the name of a new function and the sequence of statements that execute when the function is called. REUSABLE!
- The first line of the function definition is called the header; the rest is called the body. The header has to end with a colon and the body has to be indented.
- The value of named function is a function object, which has type “function”.
- Once you have defined a function, you can use it inside another function:<br>
def repeat_lyrics():<br>
print_lyrics()<br>
print_lyrics()<br>
<br>
- As you might expect, you have to create a function before you can execute it. In other words, the function definition has to be executed!!! before the first time it is called.
<br>

#### Arguments, parameters

- some functions require arguments even more than one. 
<br>
- Inside the function, the arguments are assigned to variables called parameters. Here is an example of a user-defined function that takes an argument:<br>
def print_twice(bruce):<br>
print(bruce)<br>
print(bruce)<br>
<br>
- This function assigns the argument to a parameter named bruce.  When the function is called, it prints the value of the parameter (whatever it is) twice.(prev. functions def.)<br>
print_twice('Spam')<br>
Spam<br>
Spam<br>
print_twice(17)<br>
17<br>
17<br>
<br>
- we can use any kind of expression as an argument for print_twice:<br>
print_twice('Spam '*4)<br>
Spam Spam Spam Spam<br>
Spam Spam Spam Spam<br>
<br>
- You can also use a variable as an argument:<br>
michael = 'Eric, the half a bee.'<br>
print_twice(michael)<br>
Eric, the half a bee.<br>
Eric, the half a bee.<br>
<br>
- The name of the variable we pass as an argument (michael) has nothing to do with the name of the parameter (bruce). It doesn’t matter what the value was called back home (in the caller); here in print_twice, we call everybody bruce.

#### Fruitful and Void Functions

- Some of the functions we are using, such as the math functions, yield results; for lack of a better name, I call them fruitful functions. Other functions, like print_twice, perform an action but don’t return a value. They are called void functions.
<br>
- When you call a fruitful function, you almost always want to do something with the result; for example, you might assign it to a variable or use it as part of an expression:<br>
x = math.cos(radians)<br>
golden = (math.sqrt(5) + 1) / 2<br>
<br>
- Always store your results!! <br>
<br>
- Void functions might display something on the screen or have some other effect, but they don’t have a return value. If you try to assign the result to a variable, you get a special value called None. --> NoneType Class<br>
<br>
- Use return 
<br>
def addtwo(a, b):<br>
added = a + b<br>
return added<br>
x = addtwo(3, 5) --> pass anything <br>
print(x)<br>
