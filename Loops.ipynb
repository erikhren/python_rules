**WHILE STATEMENT**<br>
n = 5<br>
while n > 0:<br>
    print(n)<br>
    n = n - 1<br>
print('Blastoff!')<br>
<br>
1. Evaluate the condition, yielding True or False.
2. If the condition is false, exit the while statement and continue execution at the next statement.
3. If the condition is true, execute the body and then go back to step 1.
<br>
- We call each time we execute the body of the loop an iteration. For the above loop, we would say, “It had five iterations”, which means that the body of the loop was executed five times.
<br>
- We call the variable that changes each time the loop executes and controls when the loop finishes the iteration variable. 
- If there is no iteration variable, the loop will repeat forever, resulting in an infinite loop.
<br>
**Infinite loops**<br>
- can use break to test code for results or to make sure to stop
- Sometimes you don’t know it’s time to end a loop until you get half way through the body. In that case you can write an infinite loop on purpose and then use the break statement to jump out of the loop.
<br>
For example, suppose you want to take input from the user until they type done. --> GREAT FOR USER INTERACTIONS
(break when clicked or said something) -->  (“stop when this happens”)<br>
<br>
while True:<br>
line = input('> ')<br>
if line == 'done':<br>
    break<br>
print(line)<br>
print('Done!')<br>

**For Loops**
- Sometimes we want to loop through a set of things such as a list of words, the lines in a file, or a list of numbers.
-  the for loop is looping through a known set of items so it runs through as many iterations as there are items in the set.
-  “Run the statements in the body of the for loop once for each friend in the set named friends.” for friend in friends:<br>
print('Happy New Year:', friend)<br>
- In particular, friend is the iteration variable for the for loop. The variable friend changes for each iteration of the loop and controls when the for loop completes.
<br>

#### LOOP PATTERNS<br>

These loops (while, for) are generally constructed by:<br>
- Initializing one or more variables before the loop starts
- Performing some computation on each item in the loop body, possibly changing the variables in the body of the loop
- Looking at the resulting variables when the loop completes
<br>
e.g.: count = 0<br>
for itervar in [3, 41, 12, 9, 74, 15]:<br>
count = count + 1<br>
print('Count: ', count)<br>
<br>
total = 0<br>
for itervar in [3, 41, 12, 9, 74, 15]:<br>
total = total + itervar --> THE ACCUMULATOR<br>
print('Total: ', total)<br>
