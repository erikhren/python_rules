**String Slices**<br>
- A segment of a string is called a slice. 
- Including the first, excluding the last <br>
s = 'Monty Python'<br>
print(s[0:5])<br>
Monty<br>
print(s[6:12])<br>
Python<br>
<br>
- If you omit the first index (before the colon), the slice starts at the beginning of the string. If you omit the second index, the slice goes to the end of the string:<br>
fruit = 'banana'<br>
fruit[:3]<br>
'ban'<br>
fruit[3:]<br>
'ana'<br>
<br>
- If the first index is greater than or equal to the second the result is an empty string,represented by two quotation marks:<br>
fruit = 'banana'<br>
fruit[3:3]<br>
''<br>

FORMAT operator (% --> modules operators for integers)<br>
- % --> allows us to construct strings, replacing parts of the strings with the data stored in variables.<br>
- embed a value in a sentence<br>
- %d --> formated to integer, %g format floats, %s for strings<br>
camels = 42<br>
'I have spotted %d camels.' % camels<br>
'I have spotted 42 camels.'<br>

- If there is more than one format sequence in the string, the second argument has to be a tuple
- Each format sequence is matched with an element of the tuple, in order.<br>
'In %d years I have spotted %g %s.' % (3, 0.1, 'camels')<br>
'In 3 years I have spotted 0.1 camels.'<br>
<br>
